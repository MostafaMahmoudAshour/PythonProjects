```python
import pandas as pd
```

```python
df = pd.read_excel('Customer Call List.xlsx')
df
```

## 1- Removing Duplicates

```python
df = df.drop_duplicates()
df
```

## 2. Removing Not useful Columns

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

## 3. Cleaning The Last_Name Column

```python
## First Way To clean the last Name
# df['Last_Name'] = df['Last_Name'].str.lstrip('/')
# df['Last_Name'] = df['Last_Name'].str.lstrip('...')
# df['Last_Name'] = df['Last_Name'].str.rstrip('_')

## Second way to clean the Last Name (more concise)
df['Last_Name'] = df['Last_Name'].str.strip('123._/') 
df
```

## 4. Clean The Phone_Number Column

```python
# we can use "regular expression" clean the Phone_Number column
import re

def func(x):
    return re.sub(r'[^a-zA-Z0-9]', '', str(x))
    
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: func(x)[:3] + '-' + func(x)[3:6] + '-' + func(x)[6:10])
df['Phone_Number'] = df['Phone_Number'].str.replace("nan--", '')
df['Phone_Number'] = df['Phone_Number'].str.replace("Na--", '')
df['Phone_Number']
```

```python
df
```

## 5. Cleaning the Address Column

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', n = 2 , expand = True)
df
```

## 6. Standardize the *Yes* And *No* values wity ***Y*** and ***N***

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df
```

## 7. Fill the Null values with Empty string

```python
df = df.fillna('')
df = df.replace('N/a', '')
df
```

## 8. Replace empty values in Don_Not_Contact columns with N (NO)

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('','N')
df
```

## 9. Removing Rows that donot allow Contact

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace = True)
df
```

## 10. Removing Rows that donnot have Phone_Number

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)
df

## Another way to drop Null values in Phone_Number column
# df = df.dropna(subset = 'Phone_Number', inplace = True)
```

```python
df = df.reset_index(drop=True)
df
```
